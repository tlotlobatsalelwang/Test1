```python
#SHOWING A RISE IN CONFIRMED CASES


import pandas as pd
import matplotlib.pyplot as plt

data_mock = {
    "country": ["Nigeria", "Nigeria", "South Africa", "South Africa", "Kenya", "Kenya"],
    "confirmed_cases": [70, 91, 76, 67, 66, 55],
    "fatalities": [84, 30, 85, 24, 4, 28],
    "recoveries": [77, 69, 28, 91, 25, 63]
}

df = pd.DataFrame(data_mock)

# A line plot with insights for each region
plt.figure(figsize=(12, 8))
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.plot(country_data.index, country_data['confirmed_cases'], label=country)

    # insights for each region
    for i, region in enumerate(country_data.index):
        plt.text(region, country_data['confirmed_cases'][region], f"{country_data['confirmed_cases'][region]} cases", ha='left', va='bottom')

plt.xlabel("Region Index")
plt.ylabel("Confirmed Cases")
plt.title("Rise of COVID-19 Cases across Regions in Example Countries")
plt.legend()
plt.xticks(df.index, df['country'])
plt.tight_layout()

plt.show()


```

```python
# SHOWING RECOVERY CASES

import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data_mock = {
    "country": ["Nigeria", "Nigeria", "South Africa", "South Africa", "Kenya", "Kenya"],
    "confirmed_cases": [70, 91, 76, 67, 66, 55],
    "fatalities": [84, 30, 85, 24, 4, 28],
    "recoveries": [77, 69, 28, 91, 25, 63]
}

df = pd.DataFrame(data_mock)

# recovery cases across regions for each country
plt.figure(figsize=(12, 8))
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.plot(country_data.index, country_data['recoveries'], label=country, marker='o')

plt.xlabel("Region Index")
plt.ylabel("Recovery Cases")
plt.title("Rise of COVID-19 Recovery Cases across Regions")
plt.legend()
plt.xticks(df.index, df['country'])
plt.tight_layout()

plt.show()

# Concise Summary
"""
- Nigeria: Steady rise in recovery cases with some fluctuations.
- South Africa: Fluctuating pattern, with a substantial increase in recoveries.
- Kenya: Gradual rise in recovery cases with a consistent trend.
"""

```

```python
# showing fatality cases

import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data_mock = {
    "country": ["Nigeria", "Nigeria", "South Africa", "South Africa", "Kenya", "Kenya"],
    "confirmed_cases": [70, 91, 76, 67, 66, 55],
    "fatalities": [84, 30, 85, 24, 4, 28],
    "recoveries": [77, 69, 28, 91, 25, 63]
}

df = pd.DataFrame(data_mock)

# Plot fatality cases across regions for each country
plt.figure(figsize=(12, 8))
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.plot(country_data.index, country_data['fatalities'], label=country, marker='o')

plt.xlabel("Region Index")
plt.ylabel("Fatality Cases")
plt.title("Rise of COVID-19 Fatality Cases across Regions")
plt.legend()
plt.xticks(df.index, df['country'])
plt.grid(True, linestyle='--', alpha=0.6)

# Add details for interpretation
plt.annotate("Significant fatalities in South Africa", xy=(3, 85), xytext=(4, 70),
             arrowprops=dict(facecolor='red', shrink=0.05), fontsize=8, color='red')

plt.tight_layout()
plt.show()
```
